```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados
df = pd.read_csv('mortalidade_recem_nascido_por100.csv')

# Verificar as primeiras linhas do DataFrame
print(df.head())

# Verificar os tipos de dados e informações básicas
print(df.info())

```

```python
# Mapeamento de cores para os países
unique_countries = df['geo'].unique()
colors = plt.cm.tab20(range(len(unique_countries)))  # Usar um colormap com 20 cores distintas
color_map = {country: colors[i % len(colors)] for i, country in enumerate(unique_countries)}

# Plotar o gráfico de dispersão
plt.figure(figsize=(14, 8))

for country in unique_countries:
    country_data = df[df['geo'] == country]
    plt.scatter(country_data['time'], country_data['newborn_mortality_rate_per_1000'],
                s=country_data['newborn_mortality_rate_per_1000'],  # Tamanho da bolha proporcional à taxa de mortalidade
                color=color_map[country],  # Cor da bolha
                alpha=0.6,  # Transparência
                label=country)

# Adicionar legendas e títulos
plt.title('Taxa de Mortalidade de Recém-Nascidos por País ao Longo do Tempo', fontsize=15)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Taxa de Mortalidade por 1000 Nascidos Vivos', fontsize=12)
plt.legend(title='País', bbox_to_anchor=(1.05, 1), loc='upper left')

# Mostrar o gráfico
plt.show()
```
